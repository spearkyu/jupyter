```python
import csv
import matplotlib.pyplot as plt

f = open('gj.csv')
data = csv.reader(f)
next(data)
result = []
result2 = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '25':
            result.append(float(row[-1]))
    if row[-2] != '':
        if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '25':
            result2.append(float(row[-2]))

plt.rc('font', family='Malgun Gothic')
plt.title('내 생일의 기온 변화 그래프')
plt.plot(result, 'hotpink', label='최고 기온')
plt.plot(result2, 'blue', label='최저 기온')
plt.legend()

# 가로축 눈금 설정
plt.xticks(range(0, 81, 10))  # 0부터 80까지 10 단위로 눈금 표시

# 축에 텍스트 추가
plt.ylabel('기온')
plt.xlabel('날짜')

plt.show()
f.close()
```

```python
import csv
import matplotlib.pyplot as plt

# 데이터 읽어오기
with open('gj.csv') as f:
    data = csv.reader(f)
    next(data)

    # 데이터 처리 및 추출
    result = []
    result2 = []
    for row in data:
        if row[-1] != '':
            if row[0].split('-')[1:] == ['09', '25']:
                result.append(float(row[-1]))
        if row[-2] != '':
            if row[0].split('-')[1:] == ['09', '25']:
                result2.append(float(row[-2]))

# 그래프 그리기
plt.rc('font', family='Malgun Gothic')
plt.title('내 생일의 기온 변화 그래프')
plt.plot(result, 'hotpink', label='최고 기온')
plt.plot(result2, 'blue', label='최저 기온')
plt.legend()

# 가로축 눈금 설정
years = range(1939, 2024, 10)
plt.xticks(range(len(years)), years, rotation=45)  # 연도 눈금과 라벨 설정

# 축에 텍스트 추가
plt.ylabel('기온')
plt.xlabel('연도')

plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus'] = False

f = open('gj.csv')
data = csv.reader(f)
next(data)

jen = []
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jen.append(float(row[-1]))
            
plt.hist(aug, bins = 100, color = 'r',label='month 8')
plt.hist(jen, bins = 100, color = 'blue',label='month 1')
plt.legend()
plt.show()

f.close()
```

```python
import csv
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus'] = False

f = open('gj.csv')
data = csv.reader(f)
next(data)

jan = []
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))
            
plt.boxplot([aug, jan])
plt.show()

f.close()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)
result = []

for row in data:
    if '일산동구' in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',', '')))  # 쉼표 제거 후 정수로 변환

plt.style.use('ggplot')
plt.plot(range(len(result)), result)  # x축 값을 명시적으로 지정
plt.show()
```
